Análisis de Sentimientos Preliminar:
Se define una función analizar_sentimiento que utiliza TextBlob para clasificar el sentimiento de un texto en positivo, neutro o negativo.
Se aplica esta función a la columna 'Text' del DataFrame y se crea una nueva columna llamada 'Sentimiento'.
Se muestra la distribución de sentimientos mediante un gráfico de barras.

```python
from textblob import TextBlob

# Función para realizar análisis de sentimientos
def analizar_sentimiento(texto):
    analysis = TextBlob(texto)
    # Clasificar el sentimiento en positivo, neutro o negativo
    if analysis.sentiment.polarity > 0:
        return 'Positivo'
    elif analysis.sentiment.polarity == 0:
        return 'Neutro'
    else:
        return 'Negativo'

# Aplicar el análisis de sentimientos a la columna 'Text'
df['Sentimiento'] = df['Text'].apply(analizar_sentimiento)

# Mostrar la distribución de los sentimientos
conteo_sentimientos = df['Sentimiento'].value_counts()

# Crear un gráfico de barras para mostrar la distribución
plt.figure(figsize=(8, 6))
conteo_sentimientos.plot(kind='bar', color=['green', 'gray', 'red'])
plt.title('Distribución de Sentimientos en Comentarios')
plt.xlabel('Sentimiento')
plt.ylabel('Número de Comentarios')
plt.show()

# Imprimir el conteo de sentimientos
print("Conteo de Sentimientos:")
print(conteo_sentimientos)
```

Ubicaciones de los restaurantes en un mapa por cercanías promedio, nos permite ver los restaurantes por cercanías filtrando por Estado. Será útil para futuros sistemas de recomendación.

```python
import folium

# Filtrar por un estado específico 
estado_seleccionado = 'FL'
restaurantes_estado = df[df['State'] == estado_seleccionado]

# Imprimir el estado de la selección
print(f"Restaurantes en el estado {estado_seleccionado}: {len(restaurantes_estado)}")

# Filtrar restaurantes con coordenadas válidas (no nulas)
restaurantes_estado = restaurantes_estado.dropna(subset=['Latitude', 'Longitude'])

# Verificar si hay al menos un restaurante con coordenadas válidas
if not restaurantes_estado.empty:
    # Crear un mapa centrado en las coordenadas promedio de los restaurantes
    mapa = folium.Map(location=[restaurantes_estado['Latitude'].mean(), restaurantes_estado['Longitude'].mean()], zoom_start=12)

    # Agregar marcadores para cada restaurante
    for index, restaurante in restaurantes_estado.iterrows():
        folium.Marker(
            location=[restaurante['Latitude'], restaurante['Longitude']],
            popup=restaurante['Name'],
            icon=folium.Icon(color='blue')
        ).add_to(mapa)

    # Mostrar el mapa
    mapa.save('mapa_restaurantes.html')  # Guardar el mapa como archivo HTML
    mapa
else:
    print(f"No hay restaurantes con coordenadas válidas en el estado {estado_seleccionado}.")
```

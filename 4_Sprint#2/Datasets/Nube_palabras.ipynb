## **Exploratory Data Analysis**

```python
# Libreria
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

### **Nubes de palabras**

```python
# Creacion dataframe
yelp=pd.read_csv("ETL_YELP_todos_estados.csv")
```

```python
# Visualizacion primeras cinco columnas
yelp.head(5)
```

Vemos que en la columna "Text" estan las rese√±as que dejan los clientes cuando van a un restaurant Mexicano.

```python
# Extraemos del dataframe la columna "Text"
text = yelp["Text"]
```

```python
# Creamos la nube de palabras
wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = None,
                min_font_size = 10).generate(' '.join(text))
```

```python
# Mostramos la nube de palabras
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)

plt.show()
```

Las palabras que mas se repiten en las rese√±as de los clientes son:

- Place     >>> Lugar
- Food      >>> Alimento
- Good      >>> Bien
- Order     >>> Orden
- Taco      >>> Taco (Plato mexicano)
- Deliciou  >>> Delicioso
- Service   >>> Servicio
- One       >>> Uno
- Ordered   >>> Ordenado
- Love      >>> Amor
